```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://movie.daum.net/ranking/reservation'
header = { "User-Agent": "Mozilla/5.0" } #브라우저 종류만 넣어주면됨(나는 브라우저에요 속이기)
html = requests.get(url, headers=header).text
soup = BeautifulSoup(html,'lxml')
```

```python
print(soup)
```

```python
ranking_numbers = soup.select('span.rank_num')
```

```python
movie_num = []
for num in ranking_numbers:
    movie_num.append(num.text)
print(movie_num)
```

```python
movie_names = soup.select('[data-tiara-layer="moviename"]')
```

```python
print(movie_names)
```

```python
names = []
for name in movie_names:
    names.append(name.text)

print(names)
```

```python
print(len(movie_num), len(names))
```

```python
for num, name in zip(movie_num, names):
    print(num + ': ' + name)
```

```python
image_tags = soup.select('.img_thumb')
print(image_tags)
```

```python
print(image_tags[0].get('src'))
```

```python
image_urls = []
for image_tag in image_tags:
    image_urls.append(image_tag.get('src'))

print(image_urls)
```

```python
print(len(image_urls))
```

```python
image_urls.insert(17,'none')
print(len(image_urls))
```

```python
import os  #폴더 저장 때문에 필요
def download_image(img_folder, url, image_name):
    if url != 'none':
        html_image = requests.get(url)
        image_file = open(os.path.join(img_folder, image_name+'.jpg'),'wb') # wb :write binary
        
        # 큰 이미지 같은 경우 나누어 받는게 좋음
        chunk_size = 1000000
        for chunk in html_image.iter_content(chunk_size):
            image_file.write(chunk)
        
        image_file.close()
    
```

```python
image_folder = '/Users/jungkihun/Documents/source/python/data_analysis/web_scrap/images'

# : , 이것이 이미지 이름에 있으면 파일이 안만들어질수 있다.
for url, name in zip(image_urls, movie_num):
    download_image(image_folder, url, name)
```
