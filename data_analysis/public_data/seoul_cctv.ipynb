```python
fp = np.polyfit(data_result['인구수'],data_result['총계'],1) #1차원 함수: y = ax+b  도출 array([a,b])
fp
```

linspace(100000,700000,100) 10만 부터 70만까지 100개로 쪼개기

```python
f1 = np.poly1d(fp)
fx = np.linspace(100000,700000,100)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'],data_result['총계'],s=50)
plt.plot(fx,f1(fx),ls ='dashed',color='g')
plt.xlabel('인구수')
plt.ylabel('cctv')
plt.grid()
plt.show()
```

```python
#오차가 큰것 6개 뽑기
data_result['오차'] = np.abs(data_result['총계']-f1(data_result['인구수']))
df_sort = data_result.sort_values(by='오차',ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'],data_result['총계'],c=data_result['오차'],s=50)
plt.plot(fx,f1(fx),ls='dashed',color='g')
# 첫번째 인자는 x좌표, 두번째 인자는 y좌표, 셋번째인자는 찍을 글자  , ppt만들때 사용함
for n in range(10):
   plt. text(df_sort['인구수'][n]*1.02, df_sort['총계'][n]*0.98, df_sort.index[n], fontsize=15)


plt.xlabel('인구수')
plt.ylabel('cctv')
plt.colorbar()
plt.grid()
plt.show()
```
