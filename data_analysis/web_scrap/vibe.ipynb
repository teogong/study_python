앨범 다운로드 구현

```python
import os  #폴더 저장 때문에 필요
def download_image(img_folder, url, image_name):
    if url != 'none':
        html_image = requests.get(url)
        image_file = open(os.path.join(img_folder, image_name+'.jpg'),'wb') # wb :write binary
        
        # 큰 이미지 같은 경우 나누어 받는게 좋음
        chunk_size = 1000000
        for chunk in html_image.iter_content(chunk_size):
            image_file.write(chunk)
        
        image_file.close()
    
```

```python

image_folder = '/Users/jungkihun/Documents/source/python/data_analysis/web_scrap/images/vibe'

# : , 이것이 이미지 이름에 있으면 파일이 안만들어질수 있다.
for url, name in zip(song_imgs, ranking_num):
    download_image(image_folder, url, name.text)
```
